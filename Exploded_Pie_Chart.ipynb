```python
#
# An example of how to create an 'exploded' pie chart plus labeling and a legend.
#
# Ref: https://stackoverflow.com/questions/35488666/matplotlib-pandas-pie-chart-label-mistakes
#

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
group_names = ['2-3 km', '3-5 km', '5-7 km', '7-10 km', '10-20 km', '20-50 km', 
               '50-75 km', '75-100 km', '>100 km']

counts = pd.Series([1109, 696, 353, 192, 168, 86, 74, 65, 53], 
                   index=['20-50 km', '50-75 km', '10-20 km', '75-100 km',
                          '3-5 km', '7-10 km', '5-7 km', '>100 km', '2-3 km'])
```

```python
explode = (0, 0, 0, 0.1, 0.1, 0.2, 0.3, 0.4, 0.6)

colors = ['#191970', '#001CF0', '#0038E2', '#0055D4', '#0071C6', '#008DB8', '#00AAAA',
          '#00C69C', '#00E28E', '#00FF80', ]

counts.plot(kind='pie', fontsize=17, colors=colors, explode=explode, figsize=(10,7))

# Plotting:
plt.axis('equal')
plt.ylabel('')
plt.legend(labels=counts.index, loc='upper right')
plt.show()
```
